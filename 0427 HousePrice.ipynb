```python
# 模型4：決策樹
```

```python
'''
回歸樹
1.適合散布圖為分區的資料
2.樹的結構：root、branch、leaf
3.每個節點左為True，右為False
3.葉的值為該組平均數
4.挑選值做節點得指標：各指標最小SSE
5.挑選特徵做節點的指標：各特徵最小SSE(分類資料為Entropy、Gini)
6.重複在每一個節點該組的資料作挑選
7.深度越高，越容易Overfitting
'''
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
treeModel = DecisionTreeRegressor(max_depth = 6)
treeModel.fit(X_train,Y_train)
Y_pred = treeModel.predict(X_test)
print("R Square: ", r2_score(Y_test,Y_pred))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,Y_pred)))
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(Y_test,Y_pred, s=20, alpha = 0.3)   
plt.plot([min(Y_test),max(Y_test)], [min(Y_test),max(Y_test)])
plt.title('Predicted vs. Actual Sale Price')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
```

```python
RMSE_dict={}
for i in range(1,30):
    treeModel = DecisionTreeRegressor(max_depth = i)
    treeModel.fit(X_train,Y_train)
    Y_pred = treeModel.predict(X_test)
    RMSE_dict[i] = np.sqrt(mean_squared_error(Y_test,Y_pred))
plt.figure(figsize = (10,5))
plt.plot(list(RMSE_dict.keys()),list(RMSE_dict.values()),marker='o',color = 'red')

plt.title('n depth vs. RMSE')
plt.xlabel('n depth')
plt.ylabel('RMSE')
```
